### Importamos las librerias necesarias para el análisis


```python
import pandas as pd
import numpy as np
import io
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
df = pd.read_csv(io.BytesIO(uploaded["cancerdata.csv"]))
df
```

```python
x = df.iloc[:,2:]
y = df.iloc[:,1]
```

```python
x.head()
```

```python
y.value_counts()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=4)#random_state mide el nivel de accuracy o de precisión que requerimos
modelo = DecisionTreeClassifier()
modelo.fit(x_train, y_train)
resultado = modelo.score(x_test, y_test)
print(resultado)
```

```python
#Validación Cruzada k fold
from sklearn.model_selection import KFold

modelo = DecisionTreeClassifier()
kfold_validacion = KFold(10) #De estas 10 muestras vamos a sacar el promedio
```

```python
from sklearn.model_selection import cross_val_score

resultados = cross_val_score(modelo, x, y, cv=kfold_validacion)
print(resultados)
resultados.mean()
```

```python
#A partir de esta validacion cruzada se determina que este modelo es muy efectivo y preciso. (Entre más se acerque a 1 es más preciso)
```
